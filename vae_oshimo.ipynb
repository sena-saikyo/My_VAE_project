```python
import numpy as np
import os 
from os import listdir 
from keras.utils import load_img, img_to_array

img_size = 128 
datasets_path = "./grayscale/data1210/train"

files = sorted([filename for filename in listdir(datasets_path) if not filename.startswith('.')])

datasets = np.zeros((len(files),img_size, img_size, 1))

for i in range(0, len(files)): 
    fnameform = os.path.join(datasets_path, files[i]) 
    print(fnameform) 
    datasets[i] = np.array(img_to_array(
    load_img(fnameform, target_size=(img_size,img_size), color_mode='grayscale')))/255.0
```

```python
print(datasets.shape)
```

```python
from tensorflow import keras
from keras import layers


latent_dim = 32

encoder_inputs = keras.Input(shape=(128, 128, 1))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2D(128, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2D(256, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2D(256, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(256, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var], name="encoder")
```

```python
encoder.summary()
```

```python
import tensorflow as tf

class Sampler(layers.Layer):
    def call(self, z_mean, z_log_var):
        batch_size = tf.shape(z_mean)[0]
        z_size = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch_size, z_size))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
```

```python
latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(4 * 4 * 256, activation="relu")(latent_inputs)
x = layers.Reshape((4, 4, 256))(x)
x = layers.Conv2DTranspose(256, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(256, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(128, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
decoder_outputs = layers.Conv2D(1, 3, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
```

```python
decoder.summary()
```

```python
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.sampler = Sampler()
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(
            name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [self.total_loss_tracker,
                self.reconstruction_loss_tracker,
                self.kl_loss_tracker]

    # 大谷作成　なかったから作った
    # encoderにtest画像を入れてsamplerを通ってdecoderで再構成
    def predict(self, x):
        z_mean, z_log_var= self.encoder(x)
        z_ans = self.sampler(z_mean,z_log_var)
        y = self.decoder(z_ans)
        return y

    # fit()に学習データと教師データを入れる時、data[0]に学習データ、data[1]に教師データが入る
    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var = self.encoder(data[0]) # 学習データdeta[0]
            z = self.sampler(z_mean, z_log_var)
            reconstruction = decoder(z)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data[1], reconstruction), # 教師データdata[1]
                    axis=(1, 2)
                )
            )
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            total_loss = reconstruction_loss + tf.reduce_mean(kl_loss)
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "total_loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

```python
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam(), run_eagerly=True)
```

```python
import numpy as np
import os
from os import listdir
from keras.utils import load_img, img_to_array

img_size = 128

test_datasets_path = "./grayscale/data1210/test"

files_test = sorted([filename for filename in listdir(test_datasets_path) if not filename.startswith('.')])

test_datasets = np.zeros((len(files_test),img_size, img_size, 1))

for i in range(0, len(files_test)):
    fnameform = os.path.join(test_datasets_path, files_test[i])
    test_datasets[i] = np.array(img_to_array( \
    load_img(fnameform, target_size=(img_size,img_size), color_mode='grayscale')))/255.0  
```

```python
from PIL import Image
import numpy as np

for i in range(1,101):
    testset = test_datasets[(i-1)*100:100*i]
    x_decode = vae.predict(testset)
    # 元のデータが(128, 128, 1)の場合のサンプ
    dec = np.array(x_decode)
    # 不要な次元（1チャンネル）を削除して (128, 128) に変換
    dec = np.squeeze(dec)
    # 0~255の範囲にスケーリングしてuint8に変換
    result = (dec * 255).astype(np.uint8)
    print("---------")
    for j in range(100):
        # 配列をPIL Imageオブジェクトに変換
        image = Image.fromarray(result[j])
        # BMP形式で保存
        image.save(f"./result_images/mse_result/{(i-1)*100+j:05}.bmp")
    print((i-1)*100+j)
```

```python
# test画像を20個取り出す
hajime = 50
owari = hajime+20

cifar_test = tf.image.rgb_to_grayscale(x_test.astype("float32") / 255.0)

cifar_test_resized = cifar_test[hajime:owari]
cifar_test_resized = tf.image.resize(cifar_test_resized, (img_size,img_size))
testset = test_datasets[hajime:owari]
print(len(testset))
print(testset.shape)
print(cifar_test_resized.shape)
```

PCAでの潜在空間の次元削減

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 潜在空間データをシミュレーション
# ここでは、latentが潜在空間の表現、labelが関連するラベルであると仮定
z_mean, z_log_var = encoder.predict(test_datasets)  # 潜在空間データの例
latent_space = z_mean

# PCAを適用
pca = PCA(n_components=2)
latent_pca = pca.fit_transform(latent_space)
```

```python
labels = y_test[0:10000]
```

```python
# 結果を散布図で視覚化
plt.figure(figsize=(10, 10))
for i in range(10):
    idx = labels == i  # クラスラベルを取得
    idx = idx.squeeze()
    plt.scatter(latent_pca[idx, 0], latent_pca[idx, 1],marker="${}$".format(i), s=50, label=f'class {i}')

plt.legend()
plt.title("PCA of latent_space")
plt.xlabel("main1")
plt.ylabel("main2")
plt.grid(True)
plt.show()
```

t-SNEでの潜在空間の次元削減

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
latent_tsne = tsne.fit_transform(latent_space)
```
